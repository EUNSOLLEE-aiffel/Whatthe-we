```python
from typing_extensions import Self
from ast import If
import random

# 캐릭터 클래스
class Character:
  def __init__(self, name, level=1, health=100, attack=25, defense=5):
    self.name = name
    self.level = level
    self.health = health
    self.attack = attack
    self.defense = defense

    # 체력
  def is_alive(self):
      return self.health > 0

    # 공격 당하기
  def take_damage(self, damage):
    actual_damage = damage - self.defense
    if actual_damage < 0:
      actual_damage = 0
    self.health -= actual_damage

    # 공격 하기
  def attack_target(self, target):
      damage = random.randint(1, self.attack)
      print(f"{self.name}, {target.name} 공격하여 {damage}의 데미지")
      target.take_damage(damage)

# Player 클래스
class Player(Character):
  def __init__(self, name):
    Character.__init__(self, name, level=1, health=100, attack=25, defense=5)
    self.experience = 0 # 경험치

  # 경험치 메서드
  def gain_experience(self, amount):
    self.experience += amount
    print(f"{self.name}, {amount}의 경험치 획득!")

# Monster 클래스
class Monster(Character):
  def __init__(self, name, level):
    health = random.randint(10, 30) * level
    attack = random.randint(5, 20) * level
    defense = random.randint(1, 5) * level
    Character.__init__(self, name, level, health, attack, defense)

# battle 함수
def battle(player, monster):
    while player.is_alive() and monster.is_alive():
      # player 공격
      damage = random.randint(1, player.attack) - monster.defense
      if damage > 0:
        monster.health -= damage
        print(f"{player.name}, {monster.name}에게 {damage}만큼 공격!")
        print(f"{monster.name}의 체력: {monster.health}")

      else:
        print(f"{monster.name}에게 효과가 없는듯 하다...")

      # monster 공격
      damage = random.randint(1, monster.attack) - player.defense
      if damage > 0:
        player.health -= damage
        print(f"{monster.name}, {player.name}에게 {damage}만큼 공격!")
        print(f"{player.name}의 체력: {player.health}")
      else:
        print(f"{player.name}에게 효과는 미미했다!")

    if player.is_alive():
      print(f"휴... {monster.name}를 처치했다!")
      player.gain_experience(monster.level * 20)
    else:
      print("졌다...")
      return False
    return True

# game over 후 다시 시작할지 묻는 함수
def play_again():
  while True:
    answer = input("다시 시작할까요? (y/n): ")
    if answer == "y":
      return True
    elif answer == "n":
      print("GAME OVER")
      return False
    else:
      print("잘못된 입력입니다.")

# 데코레이터 함수
def game_victory_decorator(func):
    def wrapper(*args, **kwargs):
        print("╔════════════════════════════════════════╗")
        print("║           🎮 GAME CLEAR! 🎮            ║")
        print("╠════════════════════════════════════════╣")
        func(*args, **kwargs)
        print("╠════════════════════════════════════════╣")
        print("║       👑 Victory is Yours! 👑         ║")
        print("╚════════════════════════════════════════╝")
    return wrapper
# 승리 메시지 출력 함수
@game_victory_decorator
def victory_message():
    print("모든 몬스터를 물리쳤습니다. 승리!")

# main 함수
def main():
    monster_dict = {'슬라임': 1, '고블린': 2, '오크': 3}  # 몬스터 종류

    while True:  # 게임 루프
        player_name = input("플레이어의 이름을 입력하세요: ")  # 이름 입력
        player = Player(player_name)

        for monster_name, level in monster_dict.items():
            print(f"\n{monster_name}와 전투 시작!")
            monster = Monster(monster_name, level)
            '''이 밑에 루프 부분은 인터넷의 도움을 좀 받았습니다'''
            if not battle(player, monster):  # 전투에서 패배했을 경우
                while True:  # 재시작 여부 반복해서 묻기
                    if play_again():  # 재시작 선택
                        break  # 내부 루프 종료, 외부 게임 루프 재시작
                    else:  # 게임 종료 선택
                        return
                break  # 외부 루프에서 다음 게임으로 재시작

        # 모든 몬스터를 처치한 경우
        else:  # `for` 루프가 정상적으로 끝났을 때 실행
            victory_message()  # 승리 메시지 출력
            return  # 게임 종료


# 게임 실행
main()

```
