```python
import random

class Account:
    # 클래스 변수: 생성된 계좌 객체의 개수를 저장
    account_count = 0

    def __init__(self, account_holder, balance=0):
        self.bank_name = "SC은행"  # 은행 이름 고정
        self.account_holder = account_holder  # 예금주
        self.account_number = self.generate_account_number()  # 랜덤 계좌 번호 생성
        self.balance = balance  # 잔액
        self.deposit_count = 0  # 입금 횟수 초기화
        self.deposit_history_list = []  # 입금 내역 리스트
        self.withdraw_history_list = []  # 출금 내역 리스트

        # 클래스 변수 증가
        Account.account_count += 1

    def generate_account_number(self):
        # 계좌번호 형식: 3자리-2자리-6자리
        part1 = str(random.randint(100, 999))  # 3자리
        part2 = str(random.randint(10, 99))    # 2자리
        part3 = str(random.randint(100000, 999999))  # 6자리
        return f"{part1}-{part2}-{part3}"

    @classmethod
    def get_account_num(cls):
        # 생성된 계좌의 개수를 반환하는 클래스 메서드
        return cls.account_count

    def deposit(self, amount):
        # 입금 메서드: 최소 1원 이상만 가능
        if amount < 1:
            print("입금 금액은 최소 1원 이상이어야 합니다.")
        else:
            self.balance += amount
            self.deposit_count += 1  # 입금 횟수 증가
            self.deposit_history_list.append(amount)  # 입금 내역 추가
            print(f"{amount} 원이 입금되었습니다. 현재 잔액: {self.balance:,} 원")

            # 입금 횟수가 5회에 도달하면 이자 추가
            if self.deposit_count == 5:
                interest = self.balance * 0.01  # 1% 이자 계산
                self.balance += interest
                print(f"입금 횟수가 5회에 도달했습니다. 잔고에 1% 이자가 추가되어 현재 잔액: {self.balance:,} 원")

    def withdraw(self, amount):
        # 잔고 이상으로 출금할 수 없음
        if amount < 1:
            print("출금 금액은 최소 1원 이상이어야 합니다.")
        elif amount > self.balance:
            print("출금 금액이 잔고를 초과하였습니다.")
        else:
            self.balance -= amount
            self.withdraw_history_list.append(amount)  # 출금 내역 추가
            print(f"{amount} 원이 출금되었습니다. 현재 잔액: {self.balance:,} 원")

    def deposit_history(self):
        # 입금 내역을 출력
        if not self.deposit_history_list:
            print("입금 내역이 없습니다.")
        else:
            print("입금 내역:")
            for amount in self.deposit_history_list:
                print(f"- {amount} 원")

    def withdraw_history(self):
        # 출금 내역을 출력
        if not self.withdraw_history_list:
            print("출금 내역이 없습니다.")
        else:
            print("출금 내역:")
            for amount in self.withdraw_history_list:
                print(f"- {amount} 원")

    def display_info(self):
        # 계좌 정보를 출력
        print(f"은행이름: {self.bank_name}, 예금주: {self.account_holder}, "
              f"계좌번호: {self.account_number}, 잔고: {self.balance:,} 원")


# 메인 실행 부분
if __name__ == "__main__":
    # 계좌 인스턴스를 저장할 리스트 초기화
    accounts = []

    # 사용자로부터 예금주와 초기 잔액을 입력받아 계좌 생성
    for i in range(1):  # 계좌 생성
        account_holder = input(f"계좌 {i + 1}의 예금주 이름을 입력하세요: ")
        while True:
            try:
                initial_balance = int(input(f"계좌 {i + 1}의 초기 잔액을 입력하세요: "))
                if initial_balance >= 0 :
                    break
                else :
                     print("초기 잔액은 0원 이상이어야 합니다.")
            except ValueError:
                print("잘못된 입력입니다. 숫자를 입력해주세요.")

        account = Account(account_holder, initial_balance)
        accounts.append(account)
        print(f"{account_holder}님의 계좌가 성공적으로 개설되었습니다. 계좌번호: {account.account_number}")

    current_account = accounts[0] # 첫번째 계좌 선택

    # 입금 및 출금 테스트 (선택된 계좌만)
    while True:
        action = input("입금(1), 출금(2), 정보 출력(3), 내역 출력(4), 종료(5) 중 선택하세요: ")

        if action == '1':
             while True:
                  try:
                       deposit_amount = int(input("입금할 금액을 입력하세요: "))
                       if deposit_amount >= 0 :
                         break
                       else :
                           print("입금액은 0원 이상이어야 합니다.")
                  except ValueError:
                      print("잘못된 입력입니다. 숫자를 입력해주세요.")

             current_account.deposit(deposit_amount)


        elif action == '2':
              while True:
                 try:
                     withdraw_amount = int(input("출금할 금액을 입력하세요: "))
                     if withdraw_amount >= 0:
                         break
                     else :
                         print("출금액은 0원 이상이어야 합니다.")

                 except ValueError:
                    print("잘못된 입력입니다. 숫자를 입력해주세요.")
              current_account.withdraw(withdraw_amount)


        elif action == '3':
            current_account.display_info()

        elif action == '4':  # 추가된 내역 출력 기능
            print("\n[입금 내역]")
            current_account.deposit_history()  # 입금 내역 출력

            print("\n[출금 내역]")
            current_account.withdraw_history()  # 출금 내역 출력

        elif action == '5':
            break

        else:
           print("잘못된 선택입니다. 다시 선택해주세요.")

    # 잔고가 100만원 이상인 고객의 정보만 출력
    print("\n잔고가 100만원 이상인 고객 정보:")
    for account in accounts:
        if account.balance >= 1000000:
            account.display_info()

    # 생성된 계좌 개수 출력
    print(f"\n생성된 계좌 개수: {Account.get_account_num()}")
```
